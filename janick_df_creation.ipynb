# Merging GDP

```python
df_corruption = pd.melt(
    df_corruption,
    id_vars=['Economy ISO3', 'Economy Name', 'Indicator ID', 'Indicator', 'Attribute 1', 'Attribute 2', 'Attribute 3', 'Partner'],  # Columns to keep
    var_name='Year',  # Name of the new column for years
    value_name='Value'  # Name of the new column for values
)
df_corruption = df_corruption.pivot_table(
    index=['Economy ISO3', 'Economy Name', 'Year'],  # Set the unique combinations of country and year
    columns='Indicator',  # Use the indicator to create new columns
    values='Value',  # Use the 'Value' column as the values
    aggfunc='first'  # In case there are duplicates, we'll take the first value
)

df_corruption.columns = [col for col in df_corruption.columns]

# Reset index to bring back 'Economy ISO3', 'Economy Name', and 'Year' as columns
df_corruption = df_corruption.reset_index()
```

```python
df_corruption.rename(columns={'Economy Name': 'Country'}, inplace=True)
df_corruption.rename(columns={'Corruption Perceptions Index Score': 'Corruption'}, inplace=True)
df_corruption = df_corruption.drop(columns=['Corruption Perceptions Index Rank', 'Corruption Perceptions Index Sources', 'Corruption Perceptions Index Standard Error', 'Economy ISO3'])
df_corruption['Year'] = df_corruption['Year'].astype(np.int64)
```

```python
df = pd.merge(df, df_corruption, on=['Country', 'Year'], how='left')
```

# Merging Human Development Index and Gender Inequality Index

```python
def reshape_indices_df(df):
    """
    Reshape a wide-format DataFrame with HDI and GII columns for multiple years
    into a long format with Year, HDI, and GII columns.
    
    Parameters:
    df (pandas.DataFrame): Wide-format DataFrame with columns:
        Country, hdi_YYYY, gii_YYYY for years 2016-2022
    
    Returns:
    pandas.DataFrame: Long-format DataFrame with columns:
        Country, Year, HDI, GII
    """
    # Melt HDI columns
    hdi_df = df.melt(
        id_vars=['Country'],
        value_vars=[col for col in df.columns if col.startswith('hdi_')],
        var_name='Year',
        value_name='HDI'
    )
    
    # Melt GII columns
    gii_df = df.melt(
        id_vars=['Country'],
        value_vars=[col for col in df.columns if col.startswith('gii_')],
        var_name='Year',
        value_name='GII'
    )
    
    # Extract years from the Year column
    hdi_df['Year'] = hdi_df['Year'].str.extract('(\d{4})').astype(int)
    gii_df['Year'] = gii_df['Year'].str.extract('(\d{4})').astype(int)
    
    # Merge HDI and GII DataFrames
    result = pd.merge(
        hdi_df,
        gii_df[['Country', 'Year', 'GII']],
        on=['Country', 'Year'],
        how='outer'
    )
    
    # Sort by Country and Year
    result = result.sort_values(['Country', 'Year']).reset_index(drop=True)
    
    return result
```

```python
df_indices = reshape_indices_df(df_indices)
```

```python
df = pd.merge(df, df_indices, on=['Country', 'Year'], how='left')
```
